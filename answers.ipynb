## 7. Impact of buying the most popular product on uplift

```python
train_purch = pd.read_csv('train_purch.csv', parse_dates=['transaction_datetime'])
train_purch
```

```python
train_purch.product_id.value_counts()
```

```python
most_pop = train_purch.product_id.value_counts().index[0]
most_pop
```

```python
clients_mosp = train_purch.loc[train_purch.product_id == most_pop, 
                               ['client_id']]\
    .drop_duplicates()
clients_mosp
```

```python
train_mpop = pd.read_csv('train.csv')\
    .merge(clients_mosp, how='right', on='client_id')\
    .set_index('client_id')
train_mpop
```

Control group

```python
train_mpop.loc[train_mpop.treatment_flg == 0, 'purchased'].value_counts(normalize=True)
```

Treated group

```python
train_mpop.loc[train_mpop.treatment_flg == 1, 'purchased'].value_counts(normalize=True)
```

```python
train_mpop.loc[train_mpop.treatment_flg == 1, 'purchased'].value_counts(normalize=True)[1] - train_mpop.loc[train_mpop.treatment_flg == 0, 'purchased'].value_counts(normalize=True)[1]
```
